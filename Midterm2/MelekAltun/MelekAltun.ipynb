```python
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as multi 
```

```python
def ucr_ncic_code_RGroup (row):
   if row['ucr_ncic_code'] > 900.0 and row['ucr_ncic_code']<= 2500.0 :
      return 1
   if row['ucr_ncic_code'] > 2500.0 and row['ucr_ncic_code']<= 3500.0 :
      return 2
   if row['ucr_ncic_code'] > 3500.0 and row['ucr_ncic_code']<= 5000.0 :
      return 3
   if row['ucr_ncic_code'] > 5000.0 and row['ucr_ncic_code']<= 7000.0 :
      return 4
   if row['ucr_ncic_code'] > 7000.0 and row['ucr_ncic_code']<= 9000.0 :
      return 5
```

```python
df['ucr_ncic_codeCat'] = df.apply (lambda row: ucr_ncic_code_RGroup (row),axis=1)
```

```python
df2 = df[['district', 'ucr_ncic_codeCat']]
print(df2)
```

```python
# gözlenen sayımların olasılık tablosu
ct1=pd.crosstab(df2['ucr_ncic_codeCat'], df2['district'])
print (ct1) 
```

```python
# sütun yüzdeleri
colsum=ct1.sum(axis=0) 
colpct=ct1/colsum
print(colpct)
```

```python
import scipy.stats
```

```python
# chi-square
print ('chi-square value, p value, expected counts')
cs1= scipy.stats.chi2_contingency(ct1)
print (cs1)
```

```python
seaborn.factorplot(x= 'district', y= 'ucr_ncic_codeCat', data= df, kind= "bar", ci=None)
plt.xlabel('Districts')
plt.ylabel('Ucr ncic code')
plt.show()
```

### POST-HOC TESTS 

Here, we have 6 groups and there are 15 pairs; so the adjusted p-value is 0.05/15= 0,0033

```python
# her karşılaştırma için chi-square testi
recode_1v2 = {1: 1, 2: 2}
df2['Dist_1v2']= df2['district'].map(recode_1v2)

# gözlenen sayımlar arasındaki tablo
table_1v2=pd.crosstab(df2['ucr_ncic_codeCat'], df2['Dist_1v2'])
print (table_1v2)

# sütun yüzdeleri
colsum=table_1v2.sum(axis=0)
colpct=table_1v2/colsum
print(colpct)

print ('chi-square value, p value, expected counts')
cs_1v2= scipy.stats.chi2_contingency(table_1v2)
print (cs_1v2)
```
